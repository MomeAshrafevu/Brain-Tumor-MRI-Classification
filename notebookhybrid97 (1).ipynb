```python
# =========================
# Brain Tumor MRI — Stratified Split + MixUp + Focal + TTA (Final Fixed)
# InceptionResNetV2 | preprocess_input | AdamW (Cosine LR) + weight decay
# IMG=256 | 3-phase unfreezing | flow_from_dataframe + StratifiedKFold
# (Added) - Additional imports for plotting, metrics, ROC Curve, and XAI
# =========================

import os, glob, warnings
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support, balanced_accuracy_score
# (Added) - For ROC Curve
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

from tensorflow.keras import layers, models, applications
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import AdamW

# ✅ InceptionResNetV2 imports
from tensorflow.keras.applications.inception_resnet_v2 import InceptionResNetV2, preprocess_input as irv2_preprocess

# -------------------------\
# Reproducibility & Safety
# -------------------------\
warnings.filterwarnings("ignore")
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"
os.environ["TF_DETERMINISTIC_OPS"] = "1"
tf.keras.backend.clear_session()
tf.random.set_seed(42); np.random.seed(42)

# -------------------------\
# Paths & configs
# -------------------------\
DATA_DIR = "/kaggle/input/brain-tumor-mri-images-17-classes"
assert os.path.exists(DATA_DIR), f"Dataset not found: {DATA_DIR}"

IMG_SIZE = (256, 256)
BATCH    = 32

# -------------------------\
# Build file list → DataFrame
# -------------------------\
classes = sorted([d for d in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, d))])
all_paths, all_labels = [], []
for cls in classes:
    cls_dir = os.path.join(DATA_DIR, cls)
    for ext in ("*.png", "*.jpg", "*.jpeg", "*.bmp", "*.tif", "*.tiff"):
        for p in glob.glob(os.path.join(cls_dir, ext)):
            all_paths.append(p); all_labels.append(cls)

df = pd.DataFrame({"filepath": all_paths, "label": all_labels})
df = df.sample(frac=1.0, random_state=42).reset_index(drop=True)
print(f"✅ Total images: {len(df)} | Classes: {len(classes)}")

# -------------------------\
# Stratified split (≈80/20)
# -------------------------\
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
train_idx, val_idx = next(skf.split(df["filepath"], df["label"]))
df_train = df.iloc[train_idx].reset_index(drop=True)
df_val   = df.iloc[val_idx].reset_index(drop=True)
print(f"📦 Train: {len(df_train)} | Val: {len(df_val)}")

# -------------------------\
# Augmentations
# -------------------------\
train_gen = ImageDataGenerator(
    preprocessing_function=irv2_preprocess,
    rotation_range=20,
    width_shift_range=0.15, height_shift_range=0.15,
    shear_range=0.10,
    zoom_range=0.20,
    brightness_range=(0.90, 1.10),
    fill_mode='nearest'
)
val_gen = ImageDataGenerator(preprocessing_function=irv2_preprocess)

train_ds = train_gen.flow_from_dataframe(
    df_train, x_col="filepath", y_col="label",
    target_size=IMG_SIZE, class_mode="categorical",
    batch_size=BATCH, shuffle=True, seed=42
)
val_ds = val_gen.flow_from_dataframe(
    df_val, x_col="filepath", y_col="label",
    target_size=IMG_SIZE, class_mode="categorical",
    batch_size=BATCH, shuffle=False, seed=42
)

class_indices = train_ds.class_indices
num_classes   = len(class_indices)
class_names   = [k for k, _ in sorted(class_indices.items(), key=lambda x: x[1])]
print("🔤 Class indices:", class_indices)


# -----------------------------------------------
# (Added) - Dataset Visualization
# -----------------------------------------------
print("\n🖼️ --- Dataset Visualization ---")
# For denormalizing (preprocess_input scales to [-1, 1] range)
def denormalize(img_array):
    return (img_array * 0.5) + 0.5

# Load a batch from the data generator
images, labels = next(train_ds)

plt.figure(figsize=(20, 12))
for i in range(12): # Show the first 12 images
    if i >= len(images):
        break
    ax = plt.subplot(3, 4, i + 1)
    # Denormalize and show the image
    plt.imshow(denormalize(images[i]))
    plt.title(class_names[np.argmax(labels[i])], fontsize=10)
    plt.axis('off')

plt.suptitle("Sample Images from Dataset", fontsize=20)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("sample_images.png")
print("✅ Sample images saved as 'sample_images.png'.")
# Reset generators for the next steps
train_ds.reset()
val_ds.reset()
# -----------------------------------------------


# -------------------------\
# Class weights → alpha (Focal)
# -------------------------\
label_to_idx = class_indices
y_train_idx = df_train["label"].map(label_to_idx).values
cls_weights_arr = compute_class_weight(class_weight="balanced",
                                       classes=np.unique(y_train_idx),
                                       y=y_train_idx)
class_weights = {i: float(w) for i, w in enumerate(cls_weights_arr)}
print("⚖️ Class Weights:", class_weights)

def focal_categorical(y_true, y_pred, gamma=2.0, alpha=None, eps=1e-7):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.clip_by_value(y_pred, eps, 1.0 - eps)
    ce = -y_true * tf.math.log(y_pred)
    if alpha is not None:
        alpha_t = tf.reduce_sum(alpha * y_true, axis=-1, keepdims=True)
        ce = alpha_t * ce
    pt = tf.reduce_sum(y_true * y_pred, axis=-1, keepdims=True)
    loss = tf.pow(1.0 - pt, gamma) * ce
    return tf.reduce_sum(loss, axis=-1)

alpha_vec = tf.constant([class_weights[i] for i in range(num_classes)], dtype=tf.float32)
loss_fn = lambda y_true, y_pred: focal_categorical(y_true, y_pred, gamma=2.0, alpha=alpha_vec)

# -------------------------\
# MixUp
# -------------------------\
def mixup(batch_x, batch_y, alpha=0.2):
    lam = np.random.beta(alpha, alpha)
    idx = np.random.permutation(batch_x.shape[0])
    x = lam * batch_x + (1 - lam) * batch_x[idx]
    y = lam * batch_y + (1 - lam) * batch_y[idx]
    return x, y

def mixup_generator(gen, alpha=0.2):
    for bx, by in gen:
        yield mixup(bx, by, alpha)

train_mix = mixup_generator(train_ds, alpha=0.2)

# -------------------------\
# ✅ Model: InceptionResNetV2
# -------------------------\
def build_model(num_classes):
    # (Important) - Store the base model in a variable for later Grad-CAM access
    base = InceptionResNetV2(include_top=False, weights="imagenet",
                             input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
    for l in base.layers[:50]: l.trainable = False
    for l in base.layers[50:]: l.trainable = True

    x = layers.GlobalAveragePooling2D()(base.output)
    x = layers.Dropout(0.50)(x)
    x = layers.Dense(512, activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.25)(x)
    out = layers.Dense(num_classes, activation="softmax")(x)
    
    # (Change) - Return both the full model and the base model
    return models.Model(base.input, out), base

model, base_model = build_model(num_classes) # (Change)
# model.summary() 

# -------------------------\
# Optimizer + Cosine LR
# -------------------------\
steps_per_epoch = len(train_ds)
cosine_lr = tf.keras.optimizers.schedules.CosineDecayRestarts(
    initial_learning_rate=1e-4,
    first_decay_steps=5 * steps_per_epoch,
    t_mul=2.0, m_mul=0.9, alpha=5e-6
)

def compile_with(opt_lr_schedule):
    opt = AdamW(learning_rate=opt_lr_schedule, weight_decay=5e-5)
    # (Note: F1/Precision/Recall are not added here due to MixUp)
    model.compile(optimizer=opt, loss=loss_fn, metrics=["accuracy"])

ckpt = "best_inceptionresnetv2_focal_mixup_cosine_stratified.h5"
cbs = [
    EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True, verbose=1),
    ModelCheckpoint(ckpt, monitor="val_loss", save_best_only=True, verbose=1)
]

# -------------------------\
# Phase 1 — warmup
# -------------------------\
print("\n🎯 PHASE 1 (warmup): Cosine LR ~1e-4")
compile_with(cosine_lr)
h1 = model.fit(
    train_mix,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_ds,
    epochs=15, callbacks=cbs, verbose=1
)

# -------------------------\
# Phase 2 — unfreeze ~60%
# -------------------------\
n = len(base_model.layers); cut = int(0.4 * n) # (Change)
for i, l in enumerate(base_model.layers): l.trainable = (i >= cut) # (Change)
print(f"🧩 Unfreezing from layer {cut}/{n} (≈60%)")

compile_with(cosine_lr)
h2 = model.fit(
    train_mix,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_ds,
    epochs=35, callbacks=cbs, verbose=1,
    initial_epoch=len(h1.history['accuracy'])
)

# -------------------------\
# Phase 3 — unfreeze ~90%
# -------------------------\
cut = int(0.1 * n)
for i, l in enumerate(base_model.layers): l.trainable = (i >= cut) # (Change)
print(f"💎 Final unfreezing from layer {cut}/{n} (≈90%)")

compile_with(cosine_lr)
h3 = model.fit(
    train_mix,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_ds,
    epochs=75, callbacks=cbs, verbose=1,
    initial_epoch=len(h1.history['accuracy']) + len(h2.history['accuracy'])
)

print("\n✅ Training done. Loading best (val_loss)…")
model.load_weights(ckpt)


# -----------------------------------------------
# (Added) - Training Curve Plotting
# -----------------------------------------------
print("\n📈 --- Training Curves (Loss & Accuracy) ---")
try:
    # Combine the history from all 3 training phases
    h1_hist = pd.DataFrame(h1.history)
    h2_hist = pd.DataFrame(h2.history)
    h3_hist = pd.DataFrame(h3.history)
    history = pd.concat([h1_hist, h2_hist, h3_hist], ignore_index=True)

    # Loss Curve
    plt.figure(figsize=(18, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history['loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # Accuracy Curve
    plt.subplot(1, 2, 2)
    plt.plot(history['accuracy'], label='Training Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig("training_curves.png")
    print("✅ Training curves saved as 'training_curves.png'.")
except Exception as e:
    print(f"Could not plot training curves: {e}")
# -----------------------------------------------


# -------------------------\
# TTA Evaluation
# -------------------------\
def predict_tta(model, gen, tta=5):
    preds = []
    for _ in range(tta):
        gen.reset()
        preds.append(model.predict(gen, verbose=0))
    return np.mean(preds, axis=0)

print("\n📊 Final Evaluation (Validation set with TTA)")
val_ds.reset()
y_prob = predict_tta(model, val_ds, tta=5)
y_pred = np.argmax(y_prob, axis=1)
y_true = val_ds.classes

final_acc = float(np.mean(y_pred == y_true))
prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
bal_acc = balanced_accuracy_score(y_true, y_pred)

print(f"🏆 VALIDATION ACCURACY: {final_acc*100:.2f}%")
print(f"   Macro Precision: {prec*100:.2f}% | Macro Recall: {rec*100:.2f}% | Macro F1: {f1*100:.2f}%")
print(f"   Balanced Accuracy: {bal_acc*100:.2f}%")

print("\n📋 Classification Report (per-class)")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix (Validation, TTA=5) — InceptionResNetV2")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.tight_layout()
plt.savefig("confusion_matrix.png")
print("✅ Confusion matrix saved as 'confusion_matrix.png'.")


# -----------------------------------------------
# (Added) - Final Metrics Bar Chart
# -----------------------------------------------
print("\n📊 --- Final Metrics Bar Chart (Per-Class) ---")
try:
    # Get per-class metrics
    precision, recall, f1, support = precision_recall_fscore_support(
        y_true, y_pred, 
        labels=range(num_classes), 
        average=None,
        zero_division=0
    )
    metrics_df = pd.DataFrame({
        'Class': class_names,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1
    })
    # Sort by F1-Score
    metrics_df = metrics_df.sort_values(by='F1-Score', ascending=False)
    
    # Plot the bar chart
    ax = metrics_df.plot(
        x='Class', 
        y=['Precision', 'Recall', 'F1-Score'], 
        kind='bar', 
        figsize=(20, 10), 
        grid=True,
        width=0.8,
        title='Per-Class Metrics (TTA Evaluation)'
    )
    ax.set_ylabel('Score')
    ax.set_xticklabels(metrics_df['Class'], rotation=90)
    fig = ax.get_figure()
    fig.tight_layout()
    fig.savefig("per_class_metrics.png")
    print("✅ Per-Class metrics saved as 'per_class_metrics.png'.")
except Exception as e:
    print(f"Could not plot final metrics: {e}")
# -----------------------------------------------


# -----------------------------------------------
# (Added) - Multi-Class ROC Curve
# -----------------------------------------------
print("\n📈 --- ROC Curve (Micro & Macro) ---")

# One-hot-encode the true labels
y_true_bin = label_binarize(y_true, classes=range(num_classes))
# Prediction probabilities obtained from TTA
y_prob_tta = y_prob 

# Compute FPR, TPR, and ROC AUC for all classes
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob_tta[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Micro-average ROC Curve
fpr["micro"], tpr["micro"], _ = roc_curve(y_true_bin.ravel(), y_prob_tta.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Macro-average ROC Curve
# First, interpolate all FPR on a common axis
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(num_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= num_classes
fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plot the ROC curve
plt.figure(figsize=(10, 8))
plt.plot(fpr["micro"], tpr["micro"],
         label=f'Micro-average ROC (AUC = {roc_auc["micro"]:0.4f})',
         color='deeppink', linestyle=':', linewidth=4)

plt.plot(fpr["macro"], tpr["macro"],
         label=f'Macro-average ROC (AUC = {roc_auc["macro"]:0.4f})',
         color='navy', linestyle=':', linewidth=4)

plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Chance')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f'Multi-class ROC Curve (InceptionResNetV2, TTA={5})')
plt.legend(loc="lower right")
plt.grid(True)
plt.savefig("roc_curve.png")
print(f"✅ ROC Curve saved as 'roc_curve.png'. (Micro AUC: {roc_auc['micro']:.4f}, Macro AUC: {roc_auc['macro']:.4f})")
# -----------------------------------------------


# -----------------------------------------------
# (Added) - Explainable AI (Grad-CAM)
# -----------------------------------------------
print("\n🔥 --- Explainable AI (Grad-CAM) ---")

def get_img_array(img_path, size):
    img = load_img(img_path, target_size=size)
    array = img_to_array(img)
    array = np.expand_dims(array, axis=0)
    return array

def get_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    # Create the Grad-CAM model that takes an image and outputs two things:
    # 1. The output of the last convolutional layer (feature map)
    # 2. The final model prediction
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        # Call the model
        last_conv_layer_output, preds = grad_model(img_array)
        
        if pred_index is None:
            pred_index = tf.argmax(preds[0]) # Take the class with the highest probability
        
        class_channel = preds[:, pred_index] # The prediction score for that class

    # Calculate the gradient of the prediction score w.r.t. the feature map
    grads = tape.gradient(class_channel, last_conv_layer_output)
    
    # Take the mean of the gradients (this is "alpha" or channel importance)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # Multiply the feature map by the gradients (Weighted Feature Map)
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap) # Reduce dimensions

    # Apply ReLU (keep only positive contributions)
    heatmap = tf.maximum(heatmap, 0)
    
    # Normalize
    heatmap /= tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def save_gradcam_overlay(img_path, heatmap, save_path, alpha=0.4):
    # Load the original image
    img = img_to_array(load_img(img_path))
    img = denormalize(img / 255.0) # Denormalize (to 0-1 range)
    
    # Resize heatmap using Tensorflow
    heatmap_resized = tf.image.resize(
        np.expand_dims(heatmap, axis=-1), # (H, W, 1)
        (img.shape[0], img.shape[1])      # (256, 256)
    )
    heatmap_resized = tf.squeeze(heatmap_resized).numpy() # (H, W)

    # Color the heatmap
    cmap = plt.get_cmap('jet')
    heatmap_colored = cmap(heatmap_resized)[:, :, :3] # RGBA to RGB

    # Create the overlay
    overlayed_img = (heatmap_colored * alpha) + (img * (1 - alpha))

    # Save the overlay
    plt.imsave(save_path, overlayed_img)

try:
    # --- Grad-CAM Setup ---
    # Find the name of the last conv layer in the base model (right before GlobalAveragePooling)
    last_conv_layer_name = base_model.layers[-1].name
    print(f"Using layer for Grad-CAM: {last_conv_layer_name}")

    # Preprocessed input (irv2_preprocess) is needed for Grad-CAM
    # as it's required for gradient calculation
    
    # Take a few samples from the validation set
    sample_paths = df_val.sample(4, random_state=42)['filepath'].tolist()
    
    plt.figure(figsize=(15, 10))
    plt.suptitle("Explainable AI (Grad-CAM) - Model Focus", fontsize=20)
    
    for i, img_path in enumerate(sample_paths):
        # 1. Load and preprocess the image
        img_array = get_img_array(img_path, size=IMG_SIZE)
        img_array_preprocessed = irv2_preprocess(img_array.copy())
        
        # 2. Create prediction and heatmap
        preds = model.predict(img_array_preprocessed)
        pred_class_idx = np.argmax(preds[0])
        pred_class_name = class_names[pred_class_idx]
        
        heatmap = get_gradcam_heatmap(img_array_preprocessed, model, last_conv_layer_name, pred_class_idx)
        
        # 3. Original image (denormalized)
        original_img = denormalize(img_array[0] / 255.0)

        # 4. Color the heatmap (without overlay)
        heatmap_colored = plt.get_cmap('jet')(heatmap)[:, :, :3]
        heatmap_resized = tf.image.resize(heatmap_colored, (IMG_SIZE[0], IMG_SIZE[1])).numpy()
        
        # 5. Overlay image
        overlay = (heatmap_resized * 0.4) + (original_img * 0.6)

        # Create subplots
        # Original Image
        ax = plt.subplot(4, 3, i*3 + 1)
        plt.imshow(original_img)
        plt.title(f"Original Image\nTrue: {df_val[df_val['filepath']==img_path]['label'].values[0]}")
        plt.axis('off')

        # Heatmap
        ax = plt.subplot(4, 3, i*3 + 2)
        plt.imshow(heatmap_resized)
        plt.title(f"Heatmap (Focus)")
        plt.axis('off')

        # Overlay
        ax = plt.subplot(4, 3, i*3 + 3)
        plt.imshow(overlay)
        plt.title(f"Overlay\nPredicted: {pred_class_name} ({preds[0][pred_class_idx]:.2f})")
        plt.axis('off')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.savefig("gradcam_examples.png")
    print("✅ Grad-CAM examples saved as 'gradcam_examples.png'.")

except Exception as e:
    print(f"Could not plot Grad-CAM: {e}")
# -----------------------------------------------


# -------------------------\
# Save final model
# -------------------------\
model.save("final_inceptionresnetv2_focal_mixup_cosine_TTA_stratified.h5")
print("\n💾 Saved: final_inceptionresnetv2_focal_mixup_cosine_TTA_stratified.h5")
```
